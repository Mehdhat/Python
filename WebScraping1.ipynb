```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL of the page you want to scrape
url = 'https://www.daraz.pk/'
```

```python
# Send a GET request to the URL
response = requests.get(url)
```

```python
response.status_code
```

```python
print(response)
```

```python
# Check if the request was successful (status code 200)
if(response.status_code == 200):
    print("Request Successful")
else:
    print("Request Failed")
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Extract the first <h1> tag
h1_tag = soup.find('h1')

# Print the <h1> tag and its text
print(h1_tag)       # Prints the full <h1> tag
print(h1_tag.text)  # Prints just the text inside the <h1> tag
```

```python
# Extract all <p> tags
p_tags = soup.find_all('p')

# Print each <p> tag and its text
for p in p_tags:
    print(p)        # Prints the full <p> tag
    print(p.text)   # Prints just the text inside the <p> tag
```

```python
# Extract all <div> tags
div_tags = soup.find_all('div')

# Print each <div> tag and its content
for div in div_tags:
    print(div)       # Prints the full <div> tag
    print(div.text)  # Prints the text inside the <div> tag
```

```python
# Extract all <meta> tags
meta_tags = soup.find_all('meta')

# Print each <meta> tag and its attributes
for meta in meta_tags:
    print(meta)             # Prints the full <meta> tag
    print(meta.attrs)       # Prints the attributes of the <meta> tag
```

```python
# Extract all <a> tags
a_tags = soup.find_all('a')

# Print each <a> tag and its href attribute
for a in a_tags:
    href = a.get('href')  # Extract the href attribute
    text = a.text         # Extract the text inside the <a> tag
    print(f'Text: {text}, Href: {href}')
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the links on the page
    links = soup.find_all('a')  # Add this line to find links

    # Print each link
    for link in links:
        href = link.get('href')
        text = link.text
        print(f'Text: {text}, URL: {href}')
else:
    print(f'Failed to retrieve the page. Status code: {response.status_code}')
```

```python
# Find all img tags
img_tags = soup.find_all('img')

# Extract the src attribute from each img tag
img_urls = [img['src'] for img in img_tags if 'src' in img.attrs]

# Print out the image URLs
for url in img_urls:
    print(url)
```

```python

import requests
from bs4 import BeautifulSoup

# Specify the URL
url = 'https://www.daraz.pk/'

# Send a GET request to fetch the raw HTML content
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all product names (this may vary based on the website's structure)
product_names = soup.find_all('span', class_='name')

# Print out each product name
for name in product_names:
    print(name.text.strip())
```
