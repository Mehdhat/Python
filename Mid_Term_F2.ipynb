# **OS library **
The os library in Python provides a way to interact with the operating system.
os.name: Returns the name of the operating system dependent module imported. For example, it will return 'posix' for Unix-like systems, 'nt' for Windows.
Here are some basic functions from the os module:

```python
import os
print(os.name)  # Example output: 'posix' or 'nt'
```

```python
#os.getcwd(): Returns the current working directory.

import os
print(os.getcwd())  # Example output: '/home/user' or 'C:\\Users\\User'
```

```python
#os.chdir(path): Changes the current working directory to the specified path.
import os
os.chdir('/home/')
```

```python
#os.listdir(path): Returns a list of the names of the entries in the directory given by path.
import os
print(os.listdir('/home/'))
```

```python
#os.mkdir(path): Creates a directory named path.

os.mkdir(r'C:\Users\Lenovo\Desktop\New folderss1')
```

```python
#os.rmdir(path): Removes (deletes) the directory named path.

os.rmdir(r'C:\Users\Lenovo\Desktop\New folderss1')
```

```python
#os.path.exists(path): Returns True if the path exists, False otherwise.

print(os.path.exists('/path/to/file_or_directory'))
```

```python
#os.path.join(path, *paths): Joins one or more path components intelligently.

path = os.path.join('/path/to', 'directory')
print(path)  # Example output: '/path/to/directory'
```

```python
#os.path.isfile(path): Returns True if the path is a file.
print(os.path.isfile('C:\\Users\\Lenovo\\Desktop\\New folderss1'))
```

```python
#os.path.isdir(path): Returns True if the path is a directory.


import os
print(os.path.isdir('/path/to/directory'))
```

# **sys library **
The sys library in Python provides access to some variables used or maintained by the Python interpreter and to functions that interact with the interpreter. It’s particularly useful for handling command-line arguments, manipulating the Python path, and managing Python runtime environment specifics.

**Introduction to sys Library**
The sys module allows you to interact with the Python runtime environment. It provides functions and variables that can help you manage the Python interpreter, handle command-line arguments, and access system-specific parameters.

```python
#Basic Functions and Variables
#sys.argv:

#Description: A list of command-line arguments passed to the script.
#Example:

import sys
print(sys.argv)  # Example output: ['script_name.py', 'arg1', 'arg2']
```

```python
#sys.exit([arg]):

#Description: Exits from Python. The optional argument is an exit status (integer) or a string. If it’s an integer, it is used as the exit status (0 is successful, other values indicate errors). If it’s a string, it is printed to stderr and the exit status is set to 1.
#Example:


#sys.exit()  # Exit the program with status 0 (success)
```

```python
print("heloo")
```

```python
#sys.platform:

#Description: A string that identifies the platform on which Python is running. For example, 'win32', 'darwin', or 'linux'.
#Example:

import sys
print(sys.platform)  # Example output: 'win32', 'darwin', 'linux'
```

```python
#sys.version:
#Description: A string containing the version number of the Python interpreter.
#Example:

import sys
print(sys.version)  # Example output: '3.9.6 (default, Jun  1 2021, 15:50:14)
```

```python
#sys.modules:
#Description: A dictionary that maps module names to modules which have already been loaded.

print(sys.modules)  # Output: Dictionary of loaded modules
```

```python
# prompt: frozen module

import sys

# Check if Python is running in a frozen environment
is_frozen = getattr(sys, 'frozen', False)

if is_frozen:
  print("Python is running in a frozen environment.")
  # In a frozen environment, you can access the executable path using:
  executable_path = sys.executable
  print("Executable path:", executable_path)
else:
  print("Python is not running in a frozen environment.")
```

```python
#sys.stdin, sys.stdout, sys.stderr:
#Description: File objects corresponding to the interpreter’s standard input, output, and error streams. These can be used to read from and write to the standard streams.
#Example:

sys.stdout.write("Hello, world!\n")  # Writes to standard output
sys.stderr.write("An error occurred!\n")  # Writes to standard error
```

```python
import sys

# Print command-line arguments
print("Command-line arguments:", sys.argv)

# Print Python version
print("Python version:", sys.version)

# Print platform information
print("Platform:", sys.platform)

# Modify and print the module search path
sys.path.append('/new/path/to/modules')
print("Module search path:", sys.path)

# Exit the program with a custom status
sys.exit("Exiting the script")
```
